```python
# Your code goes here
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv")
df.columns = [col.lower().replace(" ", "_") for col in df.columns]

df.rename(columns= {'st': 'state'}, inplace=True)

df.head()
```

```python
def gender_replace(df):
    df['gender'] = df['gender'].map({
        'Male': 'M', 
        'female': 'F', 
        'Femal': 'F', 
        'M': 'M', 
        'F': 'F'
    })
    return df

gender_replace(df)

df["gender"].value_counts()
```

```python
df['state'].replace('AZ', 'Arizona', inplace = True)
df['state'].replace('WA', 'Washington', inplace = True)
df['state'].replace('Cali', 'California', inplace = True)

# def renaming_states():
#    df['state'].replace('AZ', 'Arizona', inplace = True)
#    df['state'].replace('WA', 'Washington', inplace = True)
#    df['state'].replace('Cali', 'California', inplace = True)
#    return new_word
# renaming_states()

df['state'].unique()
```

```python
df['state'].replace('AZ', 'Arizona', inplace = True)
df['state'].replace('WA', 'Washington', inplace = True)
df['state'].replace('Cali', 'California', inplace = True)

# def renaming_states():
#    df['state'].replace('AZ', 'Arizona', inplace = True)
#    df['state'].replace('WA', 'Washington', inplace = True)
#    df['state'].replace('Cali', 'California', inplace = True)
#    return new_word
# renaming_states()

df['state'].unique()
```

```python
# df["customer_lifetime_value"].apply(lambda x: x.stript("%") if x == str else x) WHY IS IT NOT WORKING?
df["customer_lifetime_value"] = df["customer_lifetime_value"].str.strip("%").astype(float)
df["customer_lifetime_value"].head()
```

```python
df['vehicle_class'] = df['vehicle_class'].map({
    'Four-Door Car': 'Four-Door Car', 
    'Two-Door Car': 'Two-Door Car', 
    'SUV': 'SUV', 
    'Luxury SUV': 'Luxury', 
    'Sports Car': 'Luxury', 
    'Luxury Car': 'Luxury'
    })

df["vehicle_class"].unique()
```

```python
df["customer_lifetime_value"] = df["customer_lifetime_value"].apply(lambda x: float(x) if type(x) != "NaN" else x)
df.info()
```

```python
df["number_of_open_complaints"] = df["number_of_open_complaints"].str.strip("00")
df["number_of_open_complaints"] = df["number_of_open_complaints"].str.strip("1")
df["number_of_open_complaints"] = df["number_of_open_complaints"].str.strip("/")
df['state'].replace('', '1', inplace = True)
df["number_of_open_complaints"] = df["number_of_open_complaints"].apply(lambda x: float(x) if type(x) != "nan" else x)
df["number_of_open_complaints"].unique()
df
```

```python
df.dropna(thresh=5, inplace=True)
df.shape
df.isna().sum()
```
